```python
import pickle
import matplotlib.pyplot as plt
%matplotlib notebook

def unpickle(path):
    with open(path, 'rb') as f:
        runlog = pickle.load(f)
    return runlog
```

# Loading single loss log

```python
path = '../xor/logdirs/ffbp_logdir_000/runlog_0.pkl'
loss = unpickle(path)['loss_data']
l, = plt.plot(loss)
```

```python
logdir_path = '../xor/logdirs/ffbp_logdir_000/'

paths = [logdir_path+'runlog_{}.pkl'.format(i) for i in range(3)]

losses = [unpickle(path)['loss_data'] for path in paths]

for loss in losses:
    plt.plot(loss)
```
