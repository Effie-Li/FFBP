```python
import tensorflow as tf
import numpy as np
import os

def enqueue_patterns(path_to_data_file, batch_size, inp_size, targ_size, shuffle=False, buffer=None, allow_smaller_final_batch=True):
    filename_queue = tf.train.string_input_producer(string_tensor = [path_to_data_file], shuffle = False)
    reader = tf.TextLineReader(skip_header_lines=True, name='csv_reader')
    key, record_string = reader.read(filename_queue)
    defaults = [[default_val] for x in range(inp_size + targ_size)]
    defaults.insert(0,[''])
    examples = tf.decode_csv(record_string, record_defaults=defaults)
    p = tf.transpose(examples.pop(0))
    x = tf.transpose(tf.stack(examples[0:inp_size]))
    t = tf.transpose(tf.stack(examples[inp_size:inp_size + targ_size]))
    if shuffle:
        min_after_dequeue = buffer
        capacity = min_after_dequeue + 3 * batch_size
        p_batch, x_batch, t_batch = tf.train.shuffle_batch(
            tensors = [p, x, t], 
            batch_size = batch_size, 
            capacity = capacity,
            min_after_dequeue = min_after_dequeue
        )
    else:
        p_batch, x_batch, t_batch = tf.train.batch(
            tensors = [p, x, t],
            batch_size = batch_size,
            num_threads = 1,
            capacity = 32,
            enqueue_many = False,
#             shapes = ((batch_size, inp_size), (batch_size, inp_size), (batch_size, targ_size)),
            dynamic_pad = False,
            allow_smaller_final_batch = allow_smaller_final_batch
        )

    return p_batch, x_batch, t_batch

def train(sess, examples, data_len, allow_smaller_final_batch=True):
    stdout = '  > Label: \n{}\n  > Input: \n{}\n  > Target: \n{}\n'
    print('>>> TRAIN <<<')
    for i in range(data_len // batch_size + int(allow_smaller_final_batch)):
        print('step:',i)
        pat_name, inp, targ = sess.run([example for example in examples])
        print(stdout.format(pat_name, inp, targ))

def test(sess, examples, data_len):
    print('>>> TEST <<<')
    for pl, x, t in zip([example for example in examples]):
        print(pl, x, t)
```

```python
batch_size = 1
inp_size = 2
targ_size = 1
default_val = 0.0
path='../train_data_{}.txt'
num_epochs = 10
data_len = 4

examples_train = enqueue_patterns(path.format('B'), batch_size, inp_size, targ_size, shuffle=False)
examples_test = enqueue_patterns(path.format('C'), data_len, inp_size, targ_size, shuffle=False)


with tf.Session() as sess:
    sess.run([tf.global_variables_initializer(), tf.local_variables_initializer()])
  # Start populating the filename queue.
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)
    for i in range(num_epochs):
            print('EPOCH:', i)
            if i==0 or i % 5 ==0:
                test(sess, examples_test)
            train(sess, examples_train, data_len=data_len, allow_smaller_final_batch=True)
           

    coord.request_stop()
    coord.join(threads)
```
