```python
import ipywidgets as widgets
from IPython.display import display
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

def view_img(data, ind):
    img.set_data(data[ind])

class Observer(object):
    def __init__(self, fig, label):
        self.fig = fig
        self.axes = fig.get_axes()
        self.label = label
        self.fig.canvas.mpl_connect('motion_notify_event', self)

    def __call__(self, event):
        ax = event.inaxes
        if ax:
            val = ax.get_images()[0].get_cursor_data(event)
            self.update_label(val)
        else:
            self.clear_labels()

    def update_label(self, val):
        self.label.value = 'Data: {:.4f}'.format(val)

    def clear_labels(self):
        self.label.value = 'Data: None'

# ++++++++++++++++++++++++++++++++++++++++++++++++++++++

epochs = [i for i in range(0,100,10)]
data = []
for epoch in epochs:
    data.append(np.random.rand(5,5))
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
img = ax.imshow(data[0])


# ++++++++++++++++++++++++++++++++++++++++++++++++++++++
    
step = widgets.IntSlider(
    value=0,
    min=0,
    max=len(epochs) - 1,
    step=1,
)

epoch_label = widgets.HTML(
    value="Epoch: 0",
)

val_label = widgets.HTML(
    value='Data: None'
)

def on_value_change(change):
    epoch_label.value = 'Epoch: {}'.format(epochs[change['new']])

step.observe(on_value_change, names='value')
Observer(fig, val_label)


controls_dict = {'data':widgets.fixed(data), 'ind':step}
controls_panel = widgets.Box([widgets.VBox([controls_dict['ind'], epoch_label, val_label])])
widgets.interactive_output(f=view_img, controls=controls_dict)

display(controls_panel)
```
